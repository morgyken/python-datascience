```python
import numpy as np
import pandas as pd

import sys

def main():
    dicta= {}
    response = ''
    print("=================================================================")
    print("Welcome to the Student and Assessment Management System")
    print ("<A>add a student assignment marks ")
    print(" <S>search assessment marks for a student.")
    print("<Q>quit.")
    print("=================================================================")
    resp1 = input()
    
    colnames=['Student ID','Name','WAssignments','WQuizes','WMidTerm','WFinal','WTotal']
    
    dataf = pd.DataFrame(columns=colnames)

    if resp1 == "A":
        
        while (response != ('exit')):
            print("")
            name = input ('Enter Student Name')
            studentid= input('Enter Student ID')
            assignments = float(input("Enter Assignment 1 marks: "))
            quizzes = float(input("Enter Assignment 2 marks "))
            mid_term = float(input("Enter Assignment 3 marks:  "))
            final = float(input(" Enter Final Exam mark: "))  
            print("Thank You!") 
            print("The details for the student are as follows: " + name + "Are as follows")

            data1= appendMarks(studentid,name,assignments,quizzes,mid_term, final)

            #append data to the original panda data
            dataf= dataf.append(data1)

            #write data to pandas.txt file 
            dataf.to_csv(r'students.txt', header=None, index=None, sep=',', mode='a')

            #select the last attached student 
            lastStudent = dataf.tail(1)

            #output results to the screen 
            for index, row in lastStudent.iterrows():
                print('Studnet ID: ' + str(row['Student ID']))
                print('Student Name: ' + str(row['Name']))
                print(" ")
                print ("Weighted mark for Assignment 1 out of 10: " + str( row['WAssignments']))
                print ("Weighted mark for Assignment 2 out of 20: " + str(row['WQuizes']))
                print (" ")
                print ("NWeighted mark for Assignment 3 out of 30: " + str(row['WMidTerm']))
                print (" Weighted mark for the Final Exam is out of 40: " +str(row['WFinal']))
                print(" ")
                print ("The record has been successfully added to the students.txt file.")

            print('Do you want to enter details for another student (Y/N)? Y?')
            response = input("Type 'N' to exit the program, 'Y' to continue")

            if response == 'N':             
                print("Thank you, good bye!")
                break
            else:
                continue
    elif resp1 == "S":
        while (response != ('exit')):
            studentid =input("Please enter the student ID you want to search assessment marks for: ")
                        
            #search the student using ID
            data = searchStudent(colnames, studentid)
            if len(data) >= 1:
                print("A student has been found:")
            
                #output results to the screen 
                for index, row in data.iterrows():
                    print('Studnet ID: ' + str(row['Student ID']))
                    print('Student Name: ' + str(row['Name']))
                    print(" ")
                    print ("Assignment 1 Mark: " + str( row['WAssignments']))
                    print ("Assignment 2 Mark: " + str(row['WQuizes']))
                    print (" ")
                    print ("Assignment 3 mark: " + str(row['WMidTerm']))
                    print ("Final Mark: " +str(row['WFinal']))
                    print(" ")
                    print ("Total Marks out of 100: " +str(row['WTotal']))
            print("Thank You!")
            print('Do you want to search assessment marks for another student (Y/N)? N')
            response = input("Type 'N' to exit the program, 'Y' to continue")

            if response == 'N':             
                print("Thank you, good bye!")
                break
            else:
                continue
    else:
         sys.exit('Thank you, exiting System')
            

def appendMarks(studentid,name,assignments,quizzes,mid_term, final):
    w_assignments = round(assignments*0.1,2)
    w_quizzes = round(quizzes*0.2, 2)
    w_mid_term = round(mid_term*0.3,2)
    w_final = round(final*0.4, 2)

    total = w_assignments + w_quizzes + w_mid_term + w_final
    if total > 100:
        sys.exit('Invalid Data')
        
    dicta ={'Student ID': studentid, 'Name': name, 'WAssignments':w_assignments,'WQuizes':w_quizzes,'WMidTerm':w_mid_term, 'WFinal':w_final, 'WTotal': total}        
    
    dicta= pd.DataFrame([dicta])
    
    return dicta

def searchStudent(colnames, studentid):
    
    df=pd.read_csv("students.txt", sep=',',header =None,names=colnames)
    
    data1 =df.loc[df['Student ID'] == int(studentid)]
    
    return data1

if __name__ == "__main__":
    main()
   
```
