```python
# Create a scene with a house, a tree, and a sun

import turtle
import math

# Set the background color
screen = turtle.Screen()
screen.bgcolor("skyblue")

# Create our turtle
house = turtle.Turtle()
house.color("black")
house.shape("turtle")
house.speed(10)

# Define a funtion to draw and fill a rectangle with the given

def drawRectangle(tk, x, y, color):
  tk.fillcolor(color)
  tk.begin_fill()
  tk.forward(x)
  tk.left(90)
  tk.forward(y)
  tk.left(90)
  tk.forward(x)
  tk.left(90)
  tk.forward(y)
  tk.left(90)
  tk.end_fill()
  
# Define a function to draw and fill an equalateral right 
# triangle with the given hypotenuse length and color.  This
# is used to create a roof shape.
def drawTriangle(tk, length, color):
  tk.fillcolor(color)
  tk.begin_fill()
  tk.forward(length)
  tk.left(135)
  tk.forward(length / math.sqrt(2))
  tk.left(90)
  tk.forward(length / math.sqrt(2))
  tk.left(135)
  tk.end_fill()
  

# draw the side of the house
def drawParallelogram(tk, x, y, color):
  tk.fillcolor(color)
  tk.begin_fill()
  tk.left(30)
  tk.forward(x)
  tk.left(60)
  tk.forward(y)
  tk.left(120)
  tk.forward(x)
  tk.left(60)
  tk.forward(y)
  tk.left(90)
  tk.end_fill()
  
#The turtle starts in the 
# center of the circle.
def drawFourRays(k, length, radius):
  for i in range(4):
    k.penup()
    k.forward(radius)
    k.pendown()
    k.forward(length)
    k.penup()
    k.backward(length + radius)
    k.left(90)

# Draw and fill the front of the house
house.penup() 
house.goto(-150, -120)
house.pendown()
drawRectangle(house, 100, 110, "blue")

# Draw and fill the front door
house.penup()
house.goto(-120, -120)
house.pendown()
drawRectangle(house, 40, 60, "lightgreen")

# Front roof
house.penup()
house.goto(-150, -10)
house.pendown()
drawTriangle(house, 100, "magenta")

# Side of the house
house.penup()
house.goto(-50, -120)
house.pendown()
drawParallelogram(house, 60, 110, "yellow")

# Window
house.penup()
house.goto(-30, -60)
house.pendown()
drawParallelogram(house, 20, 30, "brown")

# Side roof
house.penup()
house.goto(-50, -10)
house.pendown()
house.fillcolor("orange")
house.begin_fill()
house.left(30)
house.forward(60)
house.left(105)
house.forward(100 / math.sqrt(2))
house.left(75)
house.forward(60)
house.left(105)
house.forward(100 / math.sqrt(2))
house.left(45)
house.end_fill()

# Tree base
house.penup() 
house.goto(100, -130)
house.pendown()
drawRectangle(house, 20, 40, "brown")

# Tree top
house.penup() 
house.goto(65, -90)
house.pendown()
drawTriangle(house, 90, "lightgreen")
house.penup() 
house.goto(70, -45)
house.pendown()
drawTriangle(house, 80, "lightgreen")
house.penup() 
house.goto(75, -5)
house.pendown()
drawTriangle(house, 70, "lightgreen")

# Sun
house.penup() 
house.goto(55, 110)
house.fillcolor("yellow")
house.pendown()
house.begin_fill()
house.circle(24)
house.end_fill()

# Sun rays
house.penup()
house.goto(55, 134)
house.pendown()
drawFourRays(house, 25, 24)
house.right(45)
drawFourRays(house, 15, 24)
house.left(45)

# Put down some labels
house.color("magenta")
house.penup()
house.goto(-150, 70)
house.pendown()
style = ('Courier', 19, 'italic')
house.write("House", font=style, align='center')
house.penup()
house.goto(170, -150)
house.pendown()
house.write("Tree", font=style, align='center')
#house.write("Tree", None, None, "14pt bold")
house.penup()
house.goto(130, 150)
house.pendown()
house.write("Sun", font=style, align='center')

# Bring the turtle down to the front door, and we're done!
house.penup()
house.goto(-100, -150)
house.left(90)
```
